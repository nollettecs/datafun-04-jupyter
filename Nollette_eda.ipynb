```python
# Environment setup: Install required packages and import dependencies

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Step 1 -> Data Acquisition

```python
# Data Acquisition

import seaborn as sns

# Load the Iris dataset
df = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(df.head())
```

Step 2 -> Initial Data Inspection

```python
# Display 10 rows of the DataFrame

print(df.head(10))
print(df.shape)
print(df.dtypes)
```

Step 3 -> Initial Descriptive Statistics

```python
# Descriptive Statistics
print (df.describe())
```

Step 4 -> Initial Data Distribution for Numerical Columns

```python
# Data Distribution for numerical columns

# Inspect histogram by numerical column
df['sepal_length'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

Observations:

Figure 1: This image shows a very normal distribution in the sepal length, with the average being right around 5.8

Figure 2: Shows 4 seperate images.
    1. Sepal length (same image as above)
    2. Sepal width also shows a normal distribution with the mean around 3.1
    3. Petal length has a very odd distribution with most petals in the 1-2 range, while also having a fair amount in the 3-7 range. This causes a mean of 3.8
    4. Petal width seems to be very reflective of petal length 

Step 5 -> Initial data Distribuion for Categorical Columns

```python
# Data distribution for categorical columns

# Value counts by categorical column
df['species'].value_counts()

# Value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distrubtion of {col}')
    plt.show()

# Show all plots
plt.show()
```

Observations:

Figure 3: We have 3 seperate species all with the same number of oberservative counts (50)

Step 6 -> Initial Data Transformation and Feature Engineering

```python
# Data Transformation

# Renaming a column
df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)
df.rename(columns={'sepal_width': 'Sepal Width'}, inplace=True)
df.rename(columns={'petal_length': 'Petal Length'}, inplace=True)
df.rename(columns={'petal_width': 'Petal Width'}, inplace=True)
df.rename(columns={'species': 'Species Name'}, inplace=True)

# Add a new column
df['Sepal Area']=df['Sepal Length'] * df['Sepal Width']
```

Step 7 -> Initial Visualizations

```python
# Initial Visualizations

sns.pairplot(df, hue='Species Name')
plt.show()
```

Observations:

1. This pairplot shows multiple graphs representing the relationship between the 3 species
2. According to the graphs the Veriscolor and Virginica species seem to have a lot of similarities, while the Setosa has very different values.

```python
# Initial Visualizations

# Violin plot
sns.violinplot(x='Species Name', y='Sepal Width', data=df)
plt.show()
```

Observations:

1. This graph is showing the Sepal Width distribution for each of the individual species
